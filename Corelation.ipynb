```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv(r'/content/laptop_prices.csv')
correlation_matrix = df[['Inches', 'Price_euros']].corr()
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title("Correlation Matrix Heatmap")
plt.show()

```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
df=pd.read_csv(r'/content/laptop_prices.csv')
X = df[['Inches']]
y = df['Price_euros']
model = LinearRegression()
model.fit(X, y)
print("Coefficient:", model.coef_[0])
print("Intercept:", model.intercept_)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
df = pd.read_csv(r'/content/laptop_prices.csv')
X = df[['Inches']]
y = df['Price_euros']
model = LinearRegression()
model.fit(X, y)
predictions = model.predict(X)
plt.scatter(df['Inches'], df['Price_euros'], color='blue', label="Actual Prices")
plt.plot(df['Inches'], predictions, color='red', label="Regression Line")
plt.xlabel("Screen Size (Inches)")
plt.ylabel("Price (Euros)")
plt.title("Regression Analysis of Screen Size and Price")
plt.legend()
plt.show()
```
